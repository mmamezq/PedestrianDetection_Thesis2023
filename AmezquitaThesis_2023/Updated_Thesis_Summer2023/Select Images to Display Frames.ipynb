```python
df['m1_mean'] = ''
df['m2_mean'] = ''
df['m3_mean'] = ''
df['m4_mean'] = ''
df['m5_mean'] = ''
df['m6_mean'] = ''

for id in df['instance_id']:
    df_id = df[df.instance_id == id]
    if df_id.shape[0] < 5:
        continue
    m1_acc = np.mean(df_id.m1 == df_id.label)
    m2_acc = np.mean(df_id.m2 == df_id.label)
    m3_acc = np.mean(df_id.m3 == df_id.label)
    m4_acc = np.mean(df_id.m4 == df_id.label)
    m5_acc = np.mean(df_id.m5 == df_id.label)
    m6_acc = np.mean(df_id.m5 == df_id.label)

    df.loc[df.instance_id == id,'m1_mean'] = m1_acc
    df.loc[df.instance_id == id,'m2_mean'] = m2_acc
    df.loc[df.instance_id == id,'m3_mean'] = m3_acc
    df.loc[df.instance_id == id,'m4_mean'] = m4_acc
    df.loc[df.instance_id == id,'m5_mean'] = m5_acc
    df.loc[df.instance_id == id,'m6_mean'] = m6_acc
```

```python
selected_columns = ['instance_id', 'm1_mean', 'm2_mean', 'm3_mean', 'm4_mean', 'm5_mean', 'm6_mean']
df[selected_columns] = df[selected_columns].apply(pd.to_numeric, errors='coerce')

# Group by 'instanceid' and aggregate using mean (or first)
unique_df = df[selected_columns].groupby('instance_id').mean()


# Display the resulting DataFrame
```
